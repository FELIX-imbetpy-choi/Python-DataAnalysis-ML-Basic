```python
import matplotlib.pyplot as plt
%matplotlib inline
from wordcloud import WordCloud



```

```python
# text 문서 불러서 word 분류
text = open('./data/Independence.txt').read()
wordcloud = WordCloud().generate(text)
# 워드 클라우드는 generate만 끝나면 거의 다 끝난것.



```

```python
# 빈도계산
wordcloud.words_



```

# Wordcloud 만들기

```python
plt.figure(figsize=(12,12))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off') # 측면의 그래프 수치표시를 없애준다.
plt.show()



```

```python
# font size를 지정하여 많은 단어를 보기
wordcloud = WordCloud(max_font_size=40).generate(text) # 폰트 최고사이즈 제한 40이하

plt.figure(figsize=(12,12))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.show()







```

```python
# word count를 지정하여 단어를 보기
wordcloud = WordCloud(max_words=50).generate(text) #상위 50개만 보겠다

plt.figure(figsize=(12,12))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.show()







```

```python
# back ground 변경하기
wordcloud = WordCloud(background_color='Yellow',max_words=30).generate(text) # 몇번이상 쓰인 단어를 볼것인가? (20개 이상 쓰인 단어가 이것이다.)

plt.figure(figsize=(12,12))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.show()







```

```python
# 해상도 변경!!! 표시되는 최대글자수 변경, 글자색 변경
wordcloud = WordCloud(background_color='White',width=1280, height=720,max_words=400,colormap = 'Set1',max_font_size=40).generate(text)

plt.figure(figsize=(12,12))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.margins(x=0,y=0) # 글자들 사이의 간격
plt.show()







```

```python
# 만든 word cloud 파일로 받기
wordcloud.to_file('./data/wc_independence.png')



```

# 이미지를 이용하여 워드클라우드 모양 만들기
# 마스크를 이용하여 배경 구성하기 - 이상한나라 앨리스

```python
# 이미지 확인하기
import numpy as np
from PIL import Image #이미지 보려면
from wordcloud import  STOPWORDS # 불용어 사용 - ㅋㅋㅋ나 여러 쓸데없는 데이터 뺄 수 있는 선언



```

```python
alice_mask = np.array(Image.open('./data/alice_mask.png')) # 넘파이로 불러온 이유가???
stopwords = set(STOPWORDS)
stopwords.add('said') # 불용어 사전 등록 - 등록해서 계속해서 사용할 수 있다.



```

```python
# alice이미지에서 하얀색은 255 검정색은 0 이므로 이미지를 숫자 데이터화 시킬 수 있다.
# 숫자 데이터화를 시킨다면 머신러닝을 시킬 수 있다.
# 지문 홍채 과속단속 모두 흑백으로 만들어 데이터를 숫자화 시키고 이를 분석하는 것이다.
# 얼굴인식 같은 경우 이미지에 회색 등 여러 색들의 계층을 인식하면서 3D인식이 가능한 것이다.
# 단색인 경우 옆의 칼라와 구분이 될 때 이렇게 숫자 데이터를 이용해서 워드클라우드를 원하는 모양으로 만들 수 있다.
alice_mask







```

```python
plt.figure(figsize=(8,8))
plt.imshow(alice_mask, cmap=plt.cm.gray,interpolation='bilinear') # interploation='bilinear' 은 이미지 테두리의 값을 중간으로 두어 부드럽게 보이게 한다.
plt.axis('off')
plt.show()



```

```python
# Alice File과 Mask를 이용한 배경 구성하기
text = open('./data/alice.txt').read()
stopwords.add('little') # 이렇게 빼는 stopword를 따로 관리하는 것이 좋다.
wc = WordCloud(background_color='White', max_words=2000, mask=alice_mask, stopwords=stopwords) # mask의 역할은 모양을 잡는
wc = wc.generate(text)
wc.words_







```

# color를 grey grandation으로 변경

```python
# 흑백표시?
import random
# grey함수 - 칼라값을 가져와서 레드와 블루에서 난수값을 쓰겠다는 뜻
def grey_color_func(word, font_size, position, orientation, random_state=None, **kwargs):
    return 'hsl(0,0%%,%d%%)'%random.randint(60,100)   # R(0)G(0%%)B(랜덤) 중에 끝에 B만 사용해서 흑백처리하는 것



```

```python
plt.figure(figsize=(12,12))
plt.imshow(wc.recolor(color_func=grey_color_func,random_state=3),interpolation='bilinear')
plt.axis('off')
plt.show()



```

# 스타워즈 마스크

```python
text = open('./data/a_new_hope.txt').read()
#내용 중 단어 변경 - 카운트에서 배재할땐 불용어 / 변경해야할 땐 replace로 사용
text = text.replace('HAN','Han')
text = text.replace("LUKE'S",'Luke')



```

```python
# 이미지 확인
mask = np.array(Image.open('./data/stormtrooper_mask.png'))
mask



```

```python
plt.figure(figsize=(12,12))
plt.imshow(mask, cmap=plt.cm.gray, interpolation='bilinear')
plt.axis('off')
plt.show()



```

```python
stopwords = set(STOPWORDS) # set 치면 그전까지의 것들을 지우고 새로시작
stopwords.add('int')
stopwords.add('ext')
wc = WordCloud(max_words=1000, mask=mask, stopwords=stopwords, margin=10).generate(text)



```

```python
plt.figure(figsize=(12,12))
plt.title('Star Wars')
plt.imshow(wc, cmap=plt.cm.gray, interpolation='bilinear')
plt.axis('off')
plt.show()







```

# grey color 변경

```python
plt.figure(figsize=(12,12))
plt.title('Star Wars')
plt.imshow(wc.recolor(color_func=grey_color_func,random_state=3), cmap=plt.cm.gray, interpolation='bilinear')
plt.axis('off')
plt.show()







```
