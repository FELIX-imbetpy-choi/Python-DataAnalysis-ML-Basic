```python
# 데이터 다운로드 하기 - 1
# 바로 저장하는 방식 - 따로 우리가 변경할 수 있는 부분이 없다?
# library
import urllib.request


# URL과 저장경로 - url경로에 가서 savename으로 저장/ 저장하고자 하는 이미지의 주소 복사하여 가져올 수 있음
url = "https://uta.pw/shodou/img/6/1959.png"
savename = './data/test02.png'

# 다운로드 하기
urllib.request.urlretrieve(url,savename)
print('저장되었습니다.')







```

# 데이터 다운로드 하기 - 2
인터넷에서 지정된 파일을 메모리에 올리고 저장하기


```python
import urllib.request

url = "https://uta.pw/shodou/img/6/1959.png"
savename = './data/test03.png'
# 가져와서 파일 오픈하고 읽기
mem = urllib.request.urlopen(url).read()

# 파일로 저장 - wb를 써야 바이너리 파일로 인지한다?
with open(savename, 'wb') as f:
    f.write(mem)
print('저장되었습니다.')







```

```python
html = """
<html>
    <body>
        <div id = "meigen">
            <h1> 도서관 </h1>
            <ul class = "items">
                <li> 유니티 게임 이펙트 입문서 </li>
                <li> 스위프트로 시작하는 아이폰 앱 개발 교과서 </li>
                <li> 모던 웹 사이트 디자인의 정석 </li>
            </ul>
        </div>
    </body>
</html>

"""
# 클라스 이름은 중복될 수 있다. 아이디는 중복될 수 없음!!!!!!!!!!!!

# HTML 분석하기
soup = BeautifulSoup(html,'html.parser')
print(soup)

# 필요한 부분 CSS로 추출하기
# 타이틀 부분 추출하기
h1 = soup.select_one('div#meigen > h1').string # #표시는 id라는 뜻이다. 그 밑에 h1 키를 가져와라
print('h1 = '+h1)

# 목록 가져오기 
lists = soup.select('div#meigen > ul.items > li') # ul밑의 점은 클라스 표시 - ul이 여러개 있을 수 있음 그래서 쓰는것
lists = soup.select('div#meigen > ul.items  li')
lists = soup.select('body li')
lists = soup.select('html li')
lists = soup.select('li')
# 현재는 자손이 딱 한개 밖에 없기 때문에 상위에서 모든 후손을 불러도 다 같은게 출력된다. 하지만
# 여러개의 태그가 있을 때는 이렇게 사용하면 특정 부분을 가져오지 못함 
#print(lists)

for lst in lists:
    print('li = '+lst.string)



















```

http://info.finance.naver.com/marketindex/

```python
# 미국 환율 가져오기
from bs4 import BeautifulSoup
import urllib.request as req

# HTML
url = 'http://info.finance.naver.com/marketindex/'
res = req.urlopen(url)

# HTML 분석
soup = BeautifulSoup(res, 'html.parser')

# 데이터 추출하기
price = soup.select_one('div.head_info > span.value').string #selcet_one 이기때문에 바로 .string을 쓸 수 있다.
print('USD / KRW :', price)











```

```python
# 미국, 일본, 유럽연합, 중국의 환율

# 미국 환율 가져오기
from bs4 import BeautifulSoup
import urllib.request as req

# HTML
url = 'http://info.finance.naver.com/marketindex/'
res = req.urlopen(url)

# HTML 분석
soup = BeautifulSoup(res, 'html.parser')

# 데이터 추출하기
aprice = soup.select('div.head_info > span.value') #selcet_one 이기때문에 바로 .string을 쓸 수 있다.

# select_one는 

# print('USD / KRW :', aprice)
# print('JAP / KRW :', jprice)
stop = 0
for k in aprice:
    if stop == 0:
        print('USD / KRW :',k.string)
    elif stop == 1:
        print('JAP / KRW :',k.string)
    elif stop == 2:
        print('EU / KRW :',k.string)
    elif stop == 3:
        print('CH / KRW :',k.string)
    stop = stop + 1
    
print('_'*20)
    
national=['USD / KRW','JAP / KRW ','EU / KRW','CH / KRW']
for k in range(0,4):
    print('%s : %s'%(national[k],aprice[k].string))















```
