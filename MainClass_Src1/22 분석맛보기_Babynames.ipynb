```python
# 연도별 남아, 여아 출생건수
total_births = names.pivot_table('births', index='year', columns = 'gender', aggfunc=sum)



```

```python
total_births.head(10)



```

```python
#그래프로 그리기
total_births.plot(title='Total births by gender and year')



```

```python
# 정리된 data로 csv file 만들기 - DataFrame(names를 쓰는데/ 컬럼을 바꿔서 
names2 = pd.DataFrame(names,columns=['id','name','year','gender','births'])



```

```python
names2.head()



```

```python
# csv 저장 - index False 꼭!!!! pandas에서 index를 만들어주는데 다시 파일을 만들때 자동으로 들어가서 중복됨!
names2.to_csv('./data/births_names.csv',index = False, header = True)



```
