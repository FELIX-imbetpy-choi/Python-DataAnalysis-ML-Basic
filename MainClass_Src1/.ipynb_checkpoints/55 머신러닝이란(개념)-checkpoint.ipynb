# 머신러닝 이란?
- 수 많은 데이터를 학습시켜 거기에 있는 패턴을 찾아내는 것
- 패턴을 찾으면 그 패턴을 기반으로 데이터를 분류하거나 예측가능

어느정도 규모가 있는 샘플 데이터를 입력으로 넣어 분석 시킨다.     
그리고 이때 데이터에서 일정한 규칙을 찾아 낸다.    
그리고 찾아낸 규칙을 기반으로 다른 데이터를 분류하거나 미래를 예측하는 것 입니다.    
EX ) 손글씨 문자인식, 음성 인식, 게임 전략, 의료 진단 등   

### 어떻게 규칙을 찾을까?
머신러닝에는 벡터(Vector)라는 것이 제일 중요합니다.   
벡터란 공간에서 크기와 방향을 가진것을 의미합니다.

### 특징 추출
머신러닝을 하려면 데이터가 어떤 특징을 가지고 있는지 찾고 벡터로 만들어야 합니다.   
이 같은 변환 처리를 특징추출이라 합니다.

### 회귀분석
회귀분석은 통계 용어로 Y가 연속된 값일때 Y=f(x)와 같은 모델로 나타내는 것.   
회귀에서 사용되는 가장 기본적인 모델은 Y=ax + b 입니다. 이를 선형 회귀라고 부릅니다.   
이러한 모델에서 Y를 연속 측정의 종속변수, x를 독립변수라고 합니다.   
x가 1차원이면 단순회귀라고 부르며 2차원 이상이면 다중회귀라고 부릅니다.   

### 머신러닝의 종류
(1) 지도학습
- 데이터(문제)와 함께 답을 입력합니다. 
- 문제와 답을 알려주고 새로운 문제의 답을 예측   
ex) 주가예측

(2) 비지도 학습
- 데이터는 입력하지만 답은 입력하지 않습니다.
- 다른 데이터의 규칙성을 찾습니다.   
ex) 지문인식/답이 필요없는 데이터

### 머신러닝의 응용 분야
#### 클래스 분류
특정 데이터에 레이블을 붙여 분류할 수 있습니다.   
ex) 스팸 메일 분류, 필기인식, 증권 사기 등   
#### 클러스터링
같은 유사성을 기반으로 데이터를 여러 그룹으로 나눌 수 있습니다.   
ex) 사용자의 취향을 그룹으로 묶어 사용자 취향에 맞는 광고 제공 등.   
#### 추천
특정 데이터를 기반으로 다른 데이터를 추천하는 것   
ex) 사용자가 인터넷 서점에서 구매한 책들을 기반으로 다른 책을 추천하는 것.   
#### 회귀
과거의 데이터를 기반으로 미래의 데이터를 예측하는 것   
ex) 판매에측, 주가변동   
#### 차원축소
데이터의 특성을 유지하면서 데이터의 양을 줄이는 것.   
ex) 데이터를 시각화 하거나 구조를 추출해서 용량을 줄여 계산을 빠르게 하거나 메모리 절약
