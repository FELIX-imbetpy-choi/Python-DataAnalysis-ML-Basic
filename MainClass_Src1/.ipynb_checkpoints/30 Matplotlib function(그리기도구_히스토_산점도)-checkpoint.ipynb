# matplotlib
- data 시각화 library로 pandas에서 분석된 결과를 시각화.
- line plot, bar plot, histofram, 산점도를 주로 사용됨.
- option : 현재는 2가지( inline, nbagg)

```python
#옵션사항 넣기
%matplotlib nbagg

import numpy as np
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt







```

# Line plot
# 특정한 변수 x가 변함에 따라 종속변수 Y가 함께 변화하는데 사용

```python
# Series 에 대한 line plot - series는 1차원이다. - cumsum값의 누적값
s = pd.Series(np.random.randn(10).cumsum(),index=np.arange(0,100,10))
s



```

```python
# S로 line plot 그리기 - nbagg!!! 옵션이 들어가있어서 그래프를 조절할 수 있다.! 다 그리면 파워를 꺼줘야한다
# 꺼주지 않으면 아래의 차트 그리기 명령도 이 그래프에 오버랩 된다.
s.plot()



```

```python
#  Data frame에 대한 Line plot - series는 1차원이지만 data frame은 2차원이라 달라지는 요소들 잘 보기
df = pd.DataFrame(
    np.random.randn(10,4).cumsum(axis=0),
    columns = ['A','B','C','D'],
    index = np.arange(0,100,10)
)
df



```

```python
df.plot()



```

```python
# 데이터가 이렇게 돌아가 있으면 분석하기 힘들다.
df2 = df.T
df2



```

```python
# 데이터를 알기가 힘들다.
df2.plot()
```

```python
# 특정한 한 열만 line plot으로 그리기
df['B'].plot()



```

# Bar plot
# X 가 유한개의 값만을 가질 경우 사용 - 하나의 컬럼에 대한 추이만 본다.

```python
# Series에 대한 bar plot - 문자열을 가지고 만들어 내는 것(list)!!!!
s2 = pd.Series(np.random.rand(16), index = list('abcdefghijklmnop'))
s2



```

```python
s2.plot(kind = 'bar')



```

```python
#옆으로 나오도록 함
s2.plot(kind='barh')



```

```python
# data frame에 대한 bar plot - 데이터만 제일 앞에 써주면 컬럼 인덱스는 순서 상관 없음
df2 = pd.DataFrame(np.random.rand(6,4),
                  index = ['one','two','three','fout','five','six'],
                  columns = pd.Index(['A','B','C','D'], name = 'BoxRing')
                 )
df2







```

```python
df2.plot(kind='bar')



```

```python
df2.plot(kind='barh')



```

```python
# 누적 막대 그래프
df2.plot(kind='bar', stacked = True)



```

# Histogram
# index가 필요하지 않으며 하나의 변수에 대해 변화하는 값만 있으면 됨 - 1차원배열이다.? - 0에서 1까지 200개

```python
s3 = pd.Series(np.random.normal(0,1,size = 200))
s3



```

```python
s3.hist(bins = 200)



```

# 산점도
# 독립변수 (얻은데이터) - 측정값,조사값 //x1,x2 의 관계를 알아보고자 할 때 사용. - array데이터로 산점도 그리기?

```python
#산점도
#독립변수 (얻은데이터) - 측정값,조사값 //x1,x2 의 관계를 알아보고자 할 때 사용. - array데이터로 산점도 그리기?
x1 = np.random.normal(1,1,size = (100,1))
x2 = np.random.normal(-2,4,size = (100,1))
x = np.concatenate((x1,x2), axis = 1)
x







```

```python
# numpy 데이터를 panda로 바꿔야만 시각화가 가능하다.!!!
df3 = pd.DataFrame(x,columns=['x1','x2'])
df3.head()



```

```python
# 산점도 그리기 - 산점도는 2개이상 못쓴다. - 뒤에선 파이썬에서만 3개까지 쓰는 법 배움
plt.scatter(df3['x1'],df3['x2'])



```

```python
print(df3['x1'].corr(df3['x2'])*100,'%')



```

```python
# 서로 바뀌어도 상관계수는 변하면 안된다.!
print(df3['x2'].corr(df3['x1'])*100,'%')



```
