## Dataset 구성시에 한개의 file로 구성하기는 데이터 양이 많아 
## 조각된 file 형태로 구성후 데이터 분석시 
## 하나의 dataFrame으로 합쳐서 작업
- concatenating(연결) : 단순히 dataset를 연결
- merge(병합) : index를 중심으로 dataset을 구성 ★ 머지가 중요!
- Dataframe 구성하는 방법들 잘 보기!

```python
import numpy as np
import pandas as pd




```

# merge

```python
df1 = pd.DataFrame(
    {
        "key" : list('bbacaab'),
        "data1" : range(7)
    }
)




```

```python
df2 = pd.DataFrame(
    {
        "key" : list('abd'),
        "data1" : range(3)
    }
)




```

```python
df1




```

```python
df2
```

```python
# merge - 매개변수는 똑같이 써줘야 한다. on - df1과 df2의 공통키 중심으로 data 나열 - 데이터 길이가 다를때 헛갈림 생각해봐
pd.merge(df1,df2,on = 'key')




```

```python
# default가 inner 이다 outer 하면 합집합!
pd.merge(df2,df1,on = 'key')




```

```python
# key 중심으로 합집합 - null data도 들어간다.
pd.merge(df1,df2,on = 'key', how = 'outer')




```

```python
# df1의(왼쪽)키 중심으로 나열
pd.merge(df1,df2,on = 'key', how = 'left')




```

```python
pd.merge(df1,df2,on = 'key', how = 'right')




```

```python
df1 = pd.DataFrame(
    {
        'key' : list('bbacab'),
        'data1' : range(6)
    }

)




```

```python
df2 = pd.DataFrame(
    {
        'key' : list('ababd'),
        'data1' : range(5)
    }

)




```

```python
pd.merge(df1,df2,on = 'key')




```

```python
pd.merge(df1,df2,on = 'key', how = 'left')




```

```python
# key name이 다를 경우
df3 = pd.DataFrame(
    {
        'lkey' : list('bbaccb'),
        'data1' : range(6)
    }

)




```

```python
df4 = pd.DataFrame(
    {
        'rkey' : list('abd'),
        'data1' : range(3)
    }

)




```

```python
pd.merge(df3,df4,left_on = 'lkey', right_on = 'rkey')




```

```python
# 한개는 key, 다른 것은 index 기준의 merge
left1 = pd.DataFrame(
    {
        'key' : list('abaabc'),
        'value' : range(6)
    }

)




```

```python
# 딕셔너리에서 인덱스 구성하기
right1 = pd.DataFrame(
    {
        'group_val' : [3.5,7]
    }, index = ['a','b']

)
```

```python
right1




```

```python
pd.merge(left1,right1,left_on = 'key', right_index= True)




```

```python
# 두개모두 index기준으로 merge
left2 = pd.DataFrame(
    [
        [1,2],
        [3,4],
        [5,6]
    ], index = ['a','c','e'],
    columns = ['seoul','incheon']
)




```

```python
left2




```

```python
right2 = pd.DataFrame(
    [
        [7,8],
        [9,10],
        [11,12],
        [13,14]
    ], index = ['b','c','d','e'],
    columns = ['daegu','ulsan']
)




```

```python
right2




```

```python
pd.merge(left2,right2,left_index= True, right_index= True)




```

```python
pd.merge(left2,right2,left_index= True, right_index= True, how ='outer')




```

# Data concatenating

```python
s1 = pd.Series([0,1], index = ['a','b'])
s2 = pd.Series([2,3,4], index = ['c','d','e'])
s3 = pd.Series([5,6], index = ['f','g'])




```

```python
# concat은 그냥 계속 붙여서 쓴다.
pd.concat([s1,s2,s3])




```

```python
pd.concat([s1,s2,s3], axis=1, sort = True)




```

```python
# concat은 주로 - 데이터 가공할 때 쓴다!
s4 = pd.concat([s1*5,s3])
s4




```

```python
# DataFrame에서 concatenating사용
df1 = pd.DataFrame(
    np.arange(6).reshape(3,2),
    index = ['a','b','c'],
    columns = ['one','two']
)
df1




```

```python
df2 = pd.DataFrame(
    5 + np.arange(4).reshape(2,2),
    index = ['a','c'],
    columns = ['three','four']
)
df2




ㄴ
```

```python
pd.concat([df1,df2])




```

```python
pd.concat([df1,df2],axis = 1,sort=True)




```
