```python
# 경찰서의 위도 경도 찾기
station_lng = [] # 경찰서의 위도
station_lat = [] # 경찰서의 경도

# 사용목적에 맞게 수정
import os
import sys
import urllib.request
import json
client_id = "FhouMGJbmgH2XI8LgCYa"
client_secret = "jqJ0ZpqL8f"

for name in police_address['주소']:
    encText = urllib.parse.quote(name)
    url = "https://openapi.naver.com/v1/map/geocode?query=" + encText # json 결과
    # url = "https://openapi.naver.com/v1/map/geocode.xml?query=" + encText # xml 결과
    request = urllib.request.Request(url)
    request.add_header("X-Naver-Client-Id",client_id)
    request.add_header("X-Naver-Client-Secret",client_secret)
    response = urllib.request.urlopen(request)
    rescode = response.getcode()
    if(rescode==200):
        response_body = response.read()
        #print(response_body.decode('utf-8'))
        entext = response_body.decode('utf-8')
    else:
        print("Error Code:" + rescode)
    #---------------------------
    data = json.loads(entext)
    lng = data['result']['items'][0]['point']['x']
    lat = data['result']['items'][0]['point']['y']
    station_lng.append(lng)
    station_lat.append(lat)















```

# 데이터의 정규화 (Nomalization)★★

강도와 살인은 2자릿수(10)이나 절도와 폭력은 4자릿수(10000) 입니다.

각각의 범죄 발생 건수를 범위에 놓고 비교하는 것이 편리 합니다.

각 항목의 값을 최대 1로 계산하여 그 비중 자체로 분석하는 것이 좋습니다.

Nomalize 하는 머신러닝을 사용할 것이다.!!

```python
# 머신러닝 모듈 사용하기
# sklearn(싸이키런)의 preprocessing을 사용하면 최댓값과 최소값을 비교하면서 정규화가 진행.
from sklearn import preprocessing



```

```python
# 정규화를 진행할 칼럼 설정
col = ['강도','강간','살인','절도','폭력']
# 정규화를 진행할 데이터 설정
x = crime_anal[col].values
# 최댓값과 최소값의 스케일값 가져오기 - 최대 최소값을 분석해줌
min_max_scaler = preprocessing.MinMaxScaler()
# 정규화 시킬 최종 값은 비율로 계산되기 때문에 float 설정
x_scaled = min_max_scaler.fit_transform(x.astype(float))
# 정규화된 범죄건수가 들어있는 새로운 dataframe생성
crime_anal_norm = pd.DataFrame(x_scaled,columns=col,index=crime_anal.index)
crime_anal_norm.head()







```

```python
# 기존에 있던 검거율 컬럼을 crime_anal_norm에 추가 
col2 = ['강간검거율','강도검거율','살인검거율','절도검거율','폭력검거율']
crime_anal_norm[col2] = crime_anal[col2]
crime_anal_norm[['강도','강간','살인','절도']].corr()



```

# CCTV 데이터와 합치기 - CCTV와 범죄율과 상관성 보기
구별 CCTV수와 인구수를 가져와서 범죄율과 얼마나 상관성이 있는지 확인

```python
result_CCTV = pd.read_csv('./Data/cctv_Result.csv', encoding='utf-8', index_col='구별')
crime_anal_norm[['인구수','CCTV']]=result_CCTV[['인구수','소계']]
crime_anal_norm.head()



```

```python
# 정규화된 범죄 발생건수를 합하여 범죄 컬럼을 생성
col = ['강간','강도','살인','절도','폭력']
crime_anal_norm['범죄'] = np.sum(crime_anal_norm[col],axis=1) # axis=1 은 가로방향 
crime_anal_norm.head()



```

```python
# 검거율도 통합하여 검거 컬럼 생성
col = ['강간검거율','강도검거율','살인검거율','절도검거율','폭력검거율']
crime_anal_norm['검거'] = np.sum(crime_anal_norm[col],axis=1) 
crime_anal_norm.head()



```

# heatmap을 이용한 전체 검거율 확인
