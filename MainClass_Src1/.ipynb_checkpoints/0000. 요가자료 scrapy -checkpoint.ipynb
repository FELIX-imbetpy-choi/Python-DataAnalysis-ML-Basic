```python
ylist = []
zlist = []

for i in range(1,15):
    # HTML
    url = 'https://terms.naver.com/list.nhn?cid=51034&categoryId=51034&page=1'
    page = urlopen(url)
    soup = BeautifulSoup(page, 'html.parser')
    
    a = soup.select('#content > div.list_wrap > ul > li > div.info_area > div.subject > strong > a')
    z = soup.select('#content > div.list_wrap > ul > li > div.info_area > div:nth-of-type(2) > div > span:nth-of-type(1) > em')
    
    for k in a:
        y2list = []
        y2list.append(k['href'])
        y2list.append(k.string)
        ylist.append(y2list)

        
    for s in z:
        zlist2 = []
        zlist2.append(s.string)
        zlist.append(zlist2)
```

```python
yogafire=pd.DataFrame(ylist)
yogafire2=pd.DataFrame(zlist)
yogafire
```

```python
yogafire = yogafire.loc[range(0,419,2),:]
```

```python
yogafire.to_csv('./data/yoga1.csv',index=False, encoding='euc-kr')
```
