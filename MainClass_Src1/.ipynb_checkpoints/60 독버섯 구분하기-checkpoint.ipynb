- 학습 전용 데이터를 기반으로 다수의 의사결정 트리를 구성한 후 결과확인.

# 독버섯과 관련된 데이터를 사용한 머신러닝
https://archive.ics.uci.edu/ml/datasets/Mushroom

```python
# 데이터 획득하기
import urllib.request as req
local = './data/mushroom.csv'
url = 'https://archive.ics.uci.edu/ml/machine-learning-databases/mushroom/agaricus-lepiota.data'
req.urlretrieve(url,local)
print('ok')



```

한 라인이 버섯 한 종류를 나타냄  
1번 칼럼 : p(poisonous), e(edible)   
2번 칼럼 : b(벨모양),c(원뿔),x(볼록),f(평평),k(혹),s(오목)   
자세한 내용은 사이트 참조.   

```python
# 랜덤 포레스트를 이용한 버섯 분류 - 문자로 된 자료를 숫자로 바꾸어야?
import pandas as pd
from sklearn.ensemble import RandomForestClassifier
from sklearn import metrics
from sklearn.model_selection import train_test_split



```

```python
# 데이터 읽기
mr = pd.read_csv('./data/mushroom.csv',header=None)
mr.head()



```

```python
# 데이터 내부의 기호를 숫자로 변환하기
label = []
data = []

for row_index, row in mr.iterrows():
    label.append(row.loc[0]) # 답을 가져와서 레이블 형성
    row_data = []
    for v in row.loc[1:]:
        row_data.append(ord(v))  # ord는 문자의 ASCII code 값을 return / 반대는 chr
    data.append(row_data)







```

```python
data



```

```python
# SVM(Support Vector Machine)을 이용한 분류

import pandas as pd
from sklearn import svm, metrics
from sklearn.model_selection import train_test_split # 훈련용데이터와 테스트데이터 분류 해주는 기능


train_data, test_data, train_label, test_label = train_test_split(data, label) 
# csv_data로 train_data, test_data가 나오고 csv_label로 train_label, test_label이 나온다

# Training 학습시키기
clf = RandomForestClassifier()
clf.fit(train_data, train_label)

# Test 데이터 예측하기
pre = clf.predict(test_data)

# 정답률 맞추기
ac_score = metrics.accuracy_score(test_label, pre) 
clf_report = metrics.classification_report(test_label,pre)
print('정답률 :', ac_score)
print('Report : \n', clf_report)











```
