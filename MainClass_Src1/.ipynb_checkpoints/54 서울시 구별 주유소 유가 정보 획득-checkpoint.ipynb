# 서울시 유가 정보 분석
2018-11-30일 유가 정보 획득으로 얻은 데이터를 기준으로

```python
# 가져온 excel을 합치는 방법?? 이것 모르면 수작업으로 계속.... 이런것 잘 알아놓고 검색하고 내 길을 만들것
import pandas as pd
from glob import glob # file 합치기 툴



```

```python
station_files = glob('./data/oil/지역*.xls')
station_files



```

```python
tmp_raw=[]
for file_name in station_files:
    tmp = pd.read_excel(file_name,header=2)
    tmp_raw.append(tmp)
    
station_raw = pd.concat(tmp_raw)



```

```python
# 전체 data 갯수
len(station_raw)



```

```python
station_raw.head()



```

```python
station_raw.info()



```

```python
# 필요항목 발췌
stations = pd.DataFrame(
        {
            'Oil_store' : station_raw['상호'],
            '주소' : station_raw['주소'],
            '가격' : station_raw['휘발유'],
            '셀프' : station_raw['셀프여부'],
            '상표' : station_raw['상표']
        }
)
stations.head()







```

```python
# 주유소별 자치구 항목 추가
stations['구']=[region.split()[1] for region in stations['주소']]
stations



```

```python
stations['구'].unique() # 서울특별시 특별시 같은 오류값이 들어가 있음



```

```python
# 구가 서울특별시 인것을 확인하기
stations[stations['구'] == '서울특별시']



```

```python
# 구가 서울특별시 인것을 성동구로 변환
stations.loc[stations['구'] == '서울특별시','구'] = '성동구' # 뒤에 조건 ' 구'



```

```python
# 구가 특별시 인것을 도봉구로 변환
stations.loc[stations['구'] == '특별시','구'] = '도봉구'# 뒤에 조건 ' 구'



```

```python
# 가격 정보가 '-'인것 확인
stations.loc[stations['가격'] == '-']



```

```python
# 가격 정보가 '-'인것 제거하기
stations=stations.loc[stations['가격'] != '-']



```

```python
stations



```

```python
# 가격을 실수형으로 변환하기( 엑셀에서 문자형으로 들어갔을 수도 있기 때문에 )
stations['가격']=[float(value) for value in stations['가격']]
stations.head()



```

```python
# index 재구성
stations.reset_index(inplace=True)



```

```python
# 가격이 제일 싼 Top10 구성
High_price = stations.sort_values(by='가격',ascending=True).head(10)



```

```python
# 경찰서의 위도 경도 찾기
Low_lng = [] # 경찰서의 위도
Low_lat = [] # 경찰서의 경도

# 사용목적에 맞게 수정
import os
import sys
import urllib.request
import json
client_id = "FhouMGJbmgH2XI8LgCYa"
client_secret = "jqJ0ZpqL8f"

for name in Low_price['주소']:
    encText = urllib.parse.quote(name)
    url = "https://openapi.naver.com/v1/map/geocode?query=" + encText # json 결과
    # url = "https://openapi.naver.com/v1/map/geocode.xml?query=" + encText # xml 결과
    request = urllib.request.Request(url)
    request.add_header("X-Naver-Client-Id",client_id)
    request.add_header("X-Naver-Client-Secret",client_secret)
    response = urllib.request.urlopen(request)
    rescode = response.getcode()
    if(rescode==200):
        response_body = response.read()
        #print(response_body.decode('utf-8'))
        entext = response_body.decode('utf-8')
    else:
        print("Error Code:" + rescode)
    #---------------------------
    data = json.loads(entext)
    lng = data['result']['items'][0]['point']['x']
    lat = data['result']['items'][0]['point']['y']
    Low_lng.append(lng)
    Low_lat.append(lat)















```

```python
Low_price['lng']=Low_lng



```

```python
Low_price['lat']=Low_lat



```

```python
Low_price.index = range(0,10)
Low_price



```
