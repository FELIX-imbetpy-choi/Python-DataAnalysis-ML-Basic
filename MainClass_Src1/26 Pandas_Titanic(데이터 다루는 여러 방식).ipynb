# 컬럼 추가하기 & 수정하기

```python
# 간단하게 컬럼 추가 가능! - 갯수가 맞아야 들어간다
train['Category'] = 'Titanic'
train.head()



```

```python
# 추가하거나 수정할 때 Row의 갯수가 맞아야 추가 됨!
train['Id'] = range(891)
train.head()



```

```python
# 가족수를 구하여 FamilySize란 컬럼 생성하기



```

```python
Sibnum = train['SibSp']
Parchnum = train['Parch']
train['FamilySize'] = Sibnum+Parchnum+1
train[['SibSp','Parch','FamilySize']].head()



```

```python
# Embarked가 C인 경우 Nationality_FR에 True
# 그렇지 않은 경우 Nationality_FR에 True 값 넣기
France = train['Embarked']=='C'
train['Nationality_FR'] = France
train['Nationality_UK'] = ~France
train[['Embarked','Nationality_FR','Nationality_UK']].head() # 칼럼을 여러개 지정한대로 출력할땐 [[ ]] 에 써줌.







```

```python
# loc 사용하여 칼럼 형성 및 값넣기 - true false를 넣을땐 이걸 쓰지 않아도 되지만  - 조건에 맞는 데이터를 삽입할때 많이 쓴다.!
train.loc[train['Embarked']=='C',"Nationality"]='France'
train.loc[train['Embarked']!='C',"Nationality"]='England'
train.head(3)



```

- train['Fare']가 15 미만이면 rowFare라는 컬럼 생성 후 True 아니면 False
- train['Fare']가 15 이상 30미만이면 MediumFare라는 컬럼 생성 후 True False
- train['Fare']가 30 초과이면 HighFare라는 칼럼 생성 후 ""

```python
lowFare = train['Fare']<15
MediumFare = ((train['Fare']>=15) & (train['Fare']<30))  # 숫자는 적은것 부터 써야한다?
HighFare = train['Fare']>=30

train['LowFare'] = lowFare
train['MediumFare'] = MediumFare
train['HighFare'] = HighFare
train[['Fare','LowFare','MediumFare','HighFare']].head(10)







```

- train['Fare']가 15 미만이면 FareType이라는 컬럼 생성 후 'Low'
- train['Fare']가 15 이상 30미만이면 FareType이라는 컬럼 생성 후 'Medium'

```python
lowFare = train['Fare']<15
MediumFare = ((train['Fare']>=15) & (train['Fare']<30))  # 숫자는 적은것 부터 써야한다?
HighFare = train['Fare']>=30

train.loc[lowFare,'FareType']='Low'
train.loc[MediumFare,'FareType']='Medium'
train.loc[HighFare,'FareType']='High'
train[['Fare','FareType']].head(10)







```
