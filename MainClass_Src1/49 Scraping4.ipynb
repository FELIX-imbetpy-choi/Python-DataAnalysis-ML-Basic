```python
from bs4 import BeautifulSoup
from urllib.request import urlopen
import pandas as pd



```

```python
# HTML
url = 'https://www.rottentomatoes.com/top/bestofrt/?year=2018'
page = urlopen(url)
soup = BeautifulSoup(page, 'html.parser')



```

```python
# rating
aList = soup.select('#top_movies_main > div > table > tr span.tMeterScore') # tbody 지웠음

ratingLists = []



for lst in aList:
    ratingLists.append(lst.string.strip())

print(ratingLists)







```

```python
# title
bList = soup.select('#top_movies_main > div > table > tr > td > a') # tbody 지웠음

titleLists = []



for lst in bList:
    titleLists.append(lst.string.strip())
    
print(titleLists)







```

```python
# rank
cList = soup.select('#top_movies_main > div > table > tr > td.bold') # tbody 지웠음

rankLists = []

count = 0
for lst in cList:
    rankLists.append(lst.string.strip())

print(rankLists)







```

```python
# review
dList = soup.select('#top_movies_main > div > table > tr > td.right.hidden-xs') # tbody 지웠음

reviewLists = []


for lst in dList:
    reviewLists.append(lst.string.strip())

print(reviewLists)







```

```python
# DataFrame 만들기
df = pd.DataFrame(data = rankLists, columns=['Rank'])
df['Rating'] = ratingLists
df['Title'] = titleLists
df['Review'] = reviewLists
df.head()







```

```python
# Rank
li_ranks = soup.select('#top_movies_main > div > table > tr > td.bold')
# Rating
li_ratings = soup.select('#top_movies_main > div > table > tr > td > span > span.tMeterScore')
# Title
li_titles = soup.select('#top_movies_main > div > table > tr > td > a')
# Review
li_reviews = soup.select('#top_movies_main > div > table > tr > td.right.hidden-xs')





```

```python
rankLists = []
ratingLists = []
titleLists = []
reviewLists = []



```

```python
# rank
for li in li_ranks:
    rankT = li.string.strip()
    rankT = rankT.replace('.','')
    rankLists.append(rankT)



```

```python
# rating
for li in li_ratings:
    ratingLists.append(li.string.strip())



```

```python
# title
for li in li_titles:
    titleLists.append(li.string.strip())



```

```python
# review
for li in li_reviews:
    reviewLists.append(li.string.strip())



```

```python
import time



```

```python
# genres
urlBase = 'https://www.rottentomatoes.com' 
genreLists = []
titleLists = []
urlLists = []
count1 = 0

for li in li_titles:
    titleLists.append(li.string.strip())
    workUrl = urlBase + li['href']
    genreStr =""
    pageSelect = urlopen(workUrl)
    time.sleep(2) #3초 딜레이를 준다. 클릭하고 페이지로 이동한뒤에 가져와야하기때문에.
    soupSelect = BeautifulSoup(pageSelect, 'html.parser')
    urlSelect = soupSelect.select('#mainColumn > section.panel.panel-rt.panel-box.movie_info.media \
                         > div > div > ul > li:nth-of-type(2) > div.meta-value > a')
    for ge in urlSelect:
        genreStr = genreStr + ge.string.strip() + "|"
    genreLists.append(genreStr)
    urlLists.append(workUrl)
    
    count1 +=1
    print('>>>>>',str(count1))
    print('Title :', li.string.strip())
    print('Genere :', genreStr)
    print('URL :', workUrl)
    print('------------------------------------------------------------------')
    time.sleep(1)















```

```python
# Network Speed  문제로 제대로 추출하지 못한 데이터 확인
print(genreLists)

number1=0
count1=0

for li in genreLists:
    number1+=1
    if li == '':
        count1=+1
        print(str(number1))
        
print('-'*30)
print('Row Count :',count1)











```

```python
# url 저장해 놓은 것을 이용해서 바로 들어 가게 만드듣 것

# Genres
count1 = 0

for li in genreLists:
    count1 += 1
    if li == '':
        workUrl = urlLists[count1 - 1]
        pageSelect = urlopen(workUrl)
        genreStr = "" # 장르 초기화
        time.sleep(1) 
        soupSelect = BeautifulSoup(pageSelect, 'html.parser')
        urlSelect = soupSelect.select('#mainColumn > section.panel.panel-rt.panel-box.movie_info.media > div > div > ul > li:nth-of-type(2) > div.meta-value > a')
        
        for ge in urlSelect:
            genreStr = genreStr + ge.string.strip() + ","
    
        genreLists[count1 - 1] = genreStr
    
        print('>>>>', str(count1))
        print('Genre :', genreStr)
        print('URL :', workUrl)
        print('--------------------------------------------')











```

```python
# List를 DataFrame으로 변환
df = pd.DataFrame(data = rankLists, columns=['Rank'])
df['Rating'] = ratingLists
df['Title'] = titleLists
df['Review'] = reviewLists
df['Genre'] = genreLists
df['URL'] = urlLists



```

```python
df.to_csv('./data/rottenTomatoesLink_2018.csv',encoding='utf-8',index=False)



```
