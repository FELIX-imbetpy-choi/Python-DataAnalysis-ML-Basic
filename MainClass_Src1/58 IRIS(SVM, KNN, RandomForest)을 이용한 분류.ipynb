# iris(SVM, Support Vectior machine)을 이용한 분류

```python
# SVM(support vector machine)을 이용한 분류
import pandas as pd
from sklearn import svm, metrics
from sklearn.model_selection import  train_test_split

# Data 불러오기
csv = pd.read_csv('./data/iris.csv')

# Training 과 Test 분리하기
csv_data=csv[['SepalLength','SepalWidth','PetalLength','PetalWidth']]
csv_label = csv['Name']
train_data, test_data, train_label, test_label = train_test_split(csv_data, csv_label)

# Training 데이터 학습시키기
clf = svm.SVC()
clf.fit(train_data,train_label)

# Test 테이터 예측하기 - pre에 예측값들이 들어감
pre = clf.predict(test_data)

# 정답률 구하기
ac_score = metrics.accuracy_score(test_label, pre)
print('정답률 : ',ac_score)











```

# 100번을 돌려서 그 평균값을 보기

```python
# SVM(support vector machine)을 이용한 분류
import pandas as pd
from sklearn import svm, metrics
from sklearn.model_selection import  train_test_split

# Data 불러오기
csv = pd.read_csv('./data/iris.csv')

# Training 과 Test 분리하기
csv_data=csv[['SepalLength','SepalWidth','PetalLength','PetalWidth']]
csv_label = csv['Name']
train_data, test_data, train_label, test_label = train_test_split(csv_data, csv_label) # 요기서 랜덤으로 데이터가 형성됨


#====================================================================================

# KNN을 이용한 분류!!!!!!!!!! 차이점 보기 (1개만 바꾸면 된다?)
import pandas as pd
from sklearn import metrics # SVM이 필요 없다 지금은
from sklearn.model_selection import  train_test_split
from sklearn.neighbors import KNeighborsClassifier
import numpy as np


#====================================================================================

# RandomForest을 이용한 분류!!!!!!!!!! 차이점 보기 (1개만 바꾸면 된다?)
import pandas as pd
import numpy as np
from sklearn import metrics 
from sklearn.model_selection import  train_test_split
from sklearn.ensemble import RandomForestClassifier # 모델 바꾸는 이부분만 변경됨


SVM=[]
KNN=[]
RAN=[]
for i in range(1,101):
    # Training 데이터 학습시키기
    clf = svm.SVC()
    clf.fit(train_data,train_label)
    # Test 테이터 예측하기 - pre에 예측값들이 들어감
    pre = clf.predict(test_data)
    # 정답률 구하기
    ac_score = metrics.accuracy_score(test_label, pre)
    SVM.append(ac_score)
    
    # Training 데이터 학습시키기
    clf = KNeighborsClassifier(n_neighbors=1) # !!!! 이부분 빼고 똑같다
    clf.fit(train_data,train_label)
    # Test 테이터 예측하기 - pre에 예측값들이 들어감
    pre = clf.predict(test_data)
    # 정답률 구하기
    ac_score = metrics.accuracy_score(test_label, pre)
    KNN.append(ac_score)
    
    # Training 데이터 학습시키기
    clf = RandomForestClassifier()# !!! 이부분 빼고 똑같다!!!!
    clf.fit(train_data,train_label)
    # Test 테이터 예측하기 
    pre = clf.predict(test_data)
    # 정답률 구하기
    ac_score = metrics.accuracy_score(test_label, pre)
    RAN.append(ac_score)

print('SVM : ',np.mean(SVM))
print('KNN : ',np.mean(KNN))
print('RAN : ',np.mean(RAN))































```

# Cross-Validation (교차검증)
특정데이터를 훈련전용 데이터와 테이스 전용 데이터를 분할한 뒤   
훈련 데이터를 활용하여 학습하고, 테스트 데이터로 테스트 해서 학습의 타당성을 검증.   
모델을 확인하는 근거 자료로 사용

### SVM

```python
import pandas as pd
from sklearn import svm, metrics, model_selection 

# Data 불러오기
csv = pd.read_csv('./data/iris.csv')

# Training 과 Test 분리하기
data=csv[['SepalLength','SepalWidth','PetalLength','PetalWidth']]
label = csv['Name']

# Cross Validation 하기
clf = svm.SVC()
socores = model_selection.cross_val_score(clf, data,label,cv=3) # cross_val이 여기서 랜덤으로 값을 지정해줌 cv는 정답률 갯수

print('각각의 정답률 : ',socores)
print('평균 정답률 : ',socores.mean())











```

### KNN

```python
import pandas as pd
from sklearn import svm, metrics, model_selection 
from sklearn.neighbors import KNeighborsClassifier

# Data 불러오기
csv = pd.read_csv('./data/iris.csv')

# Training 과 Test 분리하기
data=csv[['SepalLength','SepalWidth','PetalLength','PetalWidth']]
label = csv['Name']

# Cross Validation 하기
clf = KNeighborsClassifier(n_neighbors=1)
socores = model_selection.cross_val_score(clf, data,label,cv=10) # cross_val이 여기서 랜덤으로 값을 지정해줌 cv는 정답률 갯수

print('각각의 정답률 : ',socores)
print('평균 정답률 : ',socores.mean())











```

### RandomForest

```python
import pandas as pd
from sklearn import svm, metrics, model_selection 
from sklearn.ensemble import RandomForestClassifier

# Data 불러오기
csv = pd.read_csv('./data/iris.csv')

# Training 과 Test 분리하기
data=csv[['SepalLength','SepalWidth','PetalLength','PetalWidth']]
label = csv['Name']

# Cross Validation 하기
clf = RandomForestClassifier()
socores = model_selection.cross_val_score(clf, data,label,cv=10) # cross_val이 여기서 랜덤으로 값을 지정해줌 cv는 정답률 갯수

print('각각의 정답률 : ',socores)
print('평균 정답률 : ',socores.mean())











```
