# 다음 영화 예매율

http://ticket2.movie.daum.net/movie/movieranklist.aspx 

```python

# 다음 영화 예매 순위 가져오기
from bs4 import BeautifulSoup
import urllib.request as req

# HTML
url = 'http://ticket2.movie.daum.net/Movie/MovieRankList.aspx'
res = req.urlopen(url)

# HTML 분석
soup = BeautifulSoup(res, 'html.parser')

# 데이터 추출하기
aprice = soup.select('div.desc_boxthumb > strong.tit_join > a.link_g ') #selcet_one 이기때문에 바로 .string을 쓸 수 있다.

data = []
for i, lst in enumerate(aprice):
    plus = []
    plus.append(i+1)
    plus.append(lst.string.strip())
    data.append(plus)
data











```

```python
import numpy as np
import pandas as pd



```

```python
Mlist = pd.DataFrame(data)
Mlist['index'] = range(1,21)
Mlist['title'] = Mlist[1]
Mlist



```

```python
del Mlist[0]
del Mlist[1]
Mlist



```

```python
# data 저장하기
Mlist.to_csv('./data/DaumList.csv',index=False)



```
