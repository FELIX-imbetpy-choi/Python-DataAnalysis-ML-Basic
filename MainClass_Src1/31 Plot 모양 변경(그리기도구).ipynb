```python
# plot 모양 꾸미기
# line plot
axes[0,1].plot(np.random.randn(30),color = 'g', marker = 'o', linestyle = '-.')



```

```python
# 한번에 쓰기 red * --  이순서대로 쓰려면 매개변수 필요 없이 쓸 수 있다.
axes[1,0].plot(np.random.randn(30), 'r*--')



```

```python
# 눈금, 레이블, 



```

```python
fig= plt.figure()



```

```python
ax = fig.add_subplot(1,1,1)



```

```python
ax.plot(np.random.randn(1000).cumsum())



```

```python
# 수평축의 눈금을 다르게 변경
ax.set_xticks([0,250,500,750,1000])



```

```python
# 수평축의 눈금을 문자로 대체 - 원래값이 보이고 변경된 값이 보인다.
ax.set_xticklabels(['one','two','three','four','five'],rotation=30, fontsize='7')



```

```python
ax.set_yticklabels(['un','un','dew','trois','quartro'],rotation=30, fontsize='7')



```

```python
ax.set_title("Random Data set 100")



```

```python
ax.set_xlabel('Stages')
ax.set_ylabel('Values')



```

```python
# 범례표시
fig = plt.figure()
ax = fig.add_subplot(1,1,1)



```

```python
ax.plot(np.random.randn(1000).cumsum(),'k',label = 'one')
ax.plot(np.random.randn(1000).cumsum(),'b--',label = 'two')
ax.plot(np.random.randn(1000).cumsum(),'r.',label = 'three')



```

```python
# 범례 loc = 'best' 니가 알아서 제일 좋은데 걸어
ax.legend(loc='best')



```

```python
# 수평축과 수직축의 범위 변경 - 데이터를 왜곡되게 보이는 것 방지
ax.set_xlim([100,900])
ax.set_ylim([-100,100])



```
