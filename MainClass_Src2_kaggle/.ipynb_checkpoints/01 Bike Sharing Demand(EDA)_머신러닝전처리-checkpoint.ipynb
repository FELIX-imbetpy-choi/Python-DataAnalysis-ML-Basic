## Data Field

- datetime(날짜) - hourly date + timestamp     
- season(계절) -  1 = spring, 2 = summer, 3 = fall, 4 = winter    
- holiday(휴일) - whether the day is considered a holiday   
- workingday(근무날짜) - whether the day is neither a weekend nor holiday   
- weather(날씨) -    
1: Clear, Few clouds, Partly cloudy, Partly cloudy      
2: Mist + Cloudy, Mist + Broken clouds, Mist + Few clouds, Mist     
3: Light Snow, Light Rain + Thunderstorm + Scattered clouds, Light Rain + Scattered clouds     
4: Heavy Rain + Ice Pallets + Thunderstorm + Mist, Snow + Fog    
- temp(온도) - temperature in Celsius    
- atemp(체감온도) - "feels like" temperature in Celsius   
- humidity(습도) - relative humidity    
- windspeed(바람속도) - wind speed    
- casual(비회원대여횟수) - number of non-registered user rentals initiated    
- registered(회원대여횟수) - number of registered user rentals initiated    
- count(대여횟수) - number of total rentals      

- 상기 casual, registered, count는 test data에 존재하지 않으므로 feature로 사용할 수 없음.
- 모델의 측정은 RMSLE(Root Mean Squared Logarithmic Error) 방법을 사용   
- 머신 러닝의 분류와 회귀 작업중 자전거 대여량을 예측하는 문제이므로 회귀를 사용.

- 주말과 평일로 구분해 볼 필요가 있다.   
- 분과 초는 모두 0이기 때문에 의미가 없다.

```python
# 날짜와 관련 없는 항목들의 상관계수
corrMatt = train[['temp','atemp','casual','registered','humidity','windspeed','count']]
corrMatt = corrMatt.corr()
print(corrMatt)
# 쓸만한게 없으므로(상관관계가 없음) 머신러닝 실시
```

```python
# 상관계수의 Heat Map
mask = np.array(corrMatt)
mask[np.tril_indices_from(mask)] = False # array의 1의 값을 기준으로 상위 부분은 출력 제외

fig, ax = plt.subplots()
fig.set_size_inches(20, 10)
sns.heatmap(corrMatt, mask = mask, vmax = 1, square = True, annot = True)
# 데이터는 corrMatt고, mask는 mask, vmax는 세로축 최대값
```

- 온도, 습도, 풍속은 count와 거의 연관관계가 없다.   
- count와 가장 연관이 높은 것은 registered로 등록된 대여자 이지만 test에는 이 항목이 없음.   
- atemp와 temp는 0.98로 상관관계는 높지만 다중공선성 문제때문에 feature로 사용하기에는 적합하지 않을 수 있다.   

```python
# 온도, 풍속, 습도에 따른 산점도
fig, (ax1, ax2, ax3) = plt.subplots(ncols = 3)
fig.set_size_inches(12, 5)

sns.regplot(x = 'temp', y = 'count', data = train, ax = ax1)
sns.regplot(x = 'windspeed', y = 'count', data = train, ax = ax2)
sns.regplot(x = 'humidity', y = 'count', data = train, ax = ax3)
```

```python
# datetime을 년월별로 구분하기 위해 컬럼 추가
def concatenate_year_month(datetime):
    return '{0}-{1}'.format(datetime.year, datetime.month)
```

```python
train['year_month'] = train['datetime'].apply(concatenate_year_month)
print(train.shape)
train[['datetime','year_month']].head()
```

```python
# 년도별, 월별, 년월별 막대 그래프
fig, (ax1, ax2) = plt.subplots(nrows=1, ncols=2)
fig.set_size_inches(18, 4)

sns.barplot(data = train, x = 'year', y = 'count', ax = ax1)
sns.barplot(data = train, x = 'month', y = 'count', ax = ax2)

fig, ax3 = plt.subplots(nrows=1, ncols=1)
fig.set_size_inches(18, 4)

sns.barplot(data = train, x = 'year_month', y = 'count', ax = ax3)
```

- 2011년보다 2012년의 대여량이 많다.   
- 겨울보다는 여름에 대여량이 많다.   
- 2011년과 2012년의 월별 데이터를 이어보면 전체적으로 증가하는 추세이다.

```python
# train의 count에서 low fitting과 over fitting 제거로 정규분포 구성
trainWithoutOutliers = train[np.abs(train['count'] - train['count'].mean() <= (3 * train['count'].std()))]
print(train.shape)
print(trainWithoutOutliers.shape)
```

```python
# count 값의 데이터 분포도 확인
figure, axes = plt.subplots(ncols = 2, nrows = 2)
figure.set_size_inches(12, 12)

sns.distplot(train['count'], ax = axes[0][0])
stats.probplot(train['count'], dist = 'norm', fit = True, plot = axes[0][1])

sns.distplot(np.log1p(trainWithoutOutliers['count']), ax = axes[1][0]) # np.log 는 로그를 씌운것, 표준화 시킨것.
stats.probplot(np.log1p(trainWithoutOutliers['count']), dist = 'norm', fit = True, plot = axes[1][1])
```

- count변수가 표준화를 하여도 오른쪽에 치우쳐 있음.   
- 대부분의 기계학습은 종속변수가 normal이어야 정규분포를 이룸.   
- 대안으로 이상치 데이터를 제거하고 count에 log를 씌워 변경해도 정규분포는 따르지 않지만 이전 그래프보다는 좀 더 자세히 표현하고 있음.
